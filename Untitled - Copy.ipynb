```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


file_path = "C:\\Users\\Lenovo\\Desktop\\uni\\ \\FOTORESANESH.xlsx"


data = pd.read_excel(file_path)


V1 = data['V1']
I1 = data['I1']


theta = data['theta']
I2 = data['I2']

cos2_theta = np.round(np.cos(np.radians(theta))**2, 3)

valid_indices = np.isfinite(cos2_theta) & np.isfinite(I2)
cos2_theta = cos2_theta[valid_indices]
I2 = I2[valid_indices]

output_file = "output_cos2_theta.xlsx"
pd.DataFrame({'cos^2(theta)': cos2_theta}).to_excel(output_file, index=False)

def linear_fit(x, a, b):
    return a * x + b

params1, _ = curve_fit(linear_fit, V1, I1)
best_fit_I1 = linear_fit(V1, *params1)

params2, _ = curve_fit(linear_fit, cos2_theta, I2)
best_fit_I2 = linear_fit(cos2_theta, *params2)

plt.figure(figsize=(10, 5))
plt.scatter(V1, I1, label='Data Points')
plt.plot(V1, best_fit_I1, linestyle='dotted', color='r', label=f'Best Fit Line: y = {params1[0]:.3f}x + {params1[1]:.3f}')
plt.xlabel('V(v)')
plt.ylabel('I(mA)')
plt.title('Intensity vs. Voltage')
plt.legend()
plt.grid()
plt.savefig("Intensity_vs_Voltage.png") 
plt.show()
plt.close()


plt.figure(figsize=(10, 5))
plt.scatter(cos2_theta, I2, label='Data Points')
plt.plot(cos2_theta, best_fit_I2, linestyle='dotted', color='g', label=f'Best Fit Line: y = {params2[0]:.3f}x + {params2[1]:.3f}')
plt.xlabel('cos^2(theta)')
plt.ylabel('I(mA)')
plt.title('Intensity vs. cos^2(theta)')
plt.legend()
plt.grid()
plt.savefig("Intensity_vs_cos2theta.png")  
plt.show()
plt.close()
```
